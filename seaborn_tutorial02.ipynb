# we will learn histplot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
df
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
sns.histplot(df.bill_length_mm)
```

```python
#  we can also use this method

sns.histplot(x='bill_length_mm' , data=df)
```

```python
# for horizontal
sns.histplot(y='bill_length_mm' , data=df)
```

```python
# sns.set_style('darkgrid')
sns.histplot(df.bill_length_mm )
```

```python
sns.histplot(df.bill_length_mm , kde=True )
```

```python
sns.histplot(df.bill_length_mm , kde=True ,bins=100)
```

```python
sns.histplot(df.bill_length_mm , kde=True , fill=False )
```

```python
sns.histplot(x='bill_length_mm' , data=df , hue='species')
```

```python
sns.histplot(x='bill_length_mm' , data=df , hue='species' , element='step')
```

```python
sns.histplot(x='bill_length_mm' , data=df , hue='species' , element='poly')
```

```python
sns.histplot(x='bill_length_mm' , data=df , hue='species' , multiple='layer')
```

```python
sns.histplot(x='bill_length_mm' , data=df , hue='species' , multiple='stack')
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm' , data=df)
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm' , data=df , cbar=True)
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm' , data=df , cbar=True, hue='species')
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm' , data=df ,  hue='species')
```
