```python
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x="size" , y="total_bill", data=df , hue= "sex" ,style="sex", markers=["o","<"] )

plt.title("Bill data" , fontsize=15)
plt.xlabel("No of People" , fontsize=15)
plt.ylabel("Total Bill" , fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(16,9))

sns.lineplot(x="size" , y="total_bill", data=df , hue= "day" ,style="day", markers=["o","o","o","o"] )

plt.title("Bill data" , fontsize=15)
plt.xlabel("No of People" , fontsize=15)
plt.ylabel("Total Bill" , fontsize=15)
plt.show()
```
